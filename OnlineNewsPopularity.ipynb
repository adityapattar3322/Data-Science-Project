## 4.  Understanding Target Variable

```python
df['shares'].describe()
```

```python
df['shares'].median()
```

- Since, to set the threshold for deciding whether an article is popular. We should chooose the median number of of shares in the dataset.
- All the Articles with 1400 or more shares are marked to be POPULAR and less than 1400 shares are considered as UNPOPULAR 
- for which we need to create new target variable

```python
df['popularity'] = df['shares'].apply(lambda x: 0 if x<1400 else 1)
```

```python
wday = df.columns.values[27:34]
wday
```

```python
unpop = df[df['shares']<b]
pop = df[df['shares']>=b]

pop_day = pop[wday].sum().values
unpop_day = unpop[wday].sum().values

fig = plt.figure(figsize = (13,5))
plt.title("Count of Popular and Unpopular News over different day of week (Median)",fontsize = 16)

plt.bar(np.arange(len(wday)), pop_day, width=0.3, align = 'center',color='blue', label ='Popular')
plt.bar(np.arange(len(wday))-0.3, unpop_day, width=0.3, align = 'center',color='red', label ='Unpopular') 

plt.xticks(np.arange(len(wday)),wday)
plt.ylabel('COUNT',fontsize = 15)
plt.xlabel('Day of Week', fontsize=17)

plt.legend(loc = 'upper right')
plt.tight_layout()
plt.show()
```

- Best Popular Articles are usually posted on Monday and Wednesday and (little Bit of Tuesday)
- Generally weekends (Sunday and Saturday) are the worst days to publish an article

```python
X = df.drop(['popularity'],axis=1)
y = df['popularity']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 2)
```
